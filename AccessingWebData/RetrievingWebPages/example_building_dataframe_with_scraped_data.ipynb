# Analysing corey's blog's articles

```python
# importing libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Scraping some data

#webpage = requests.get('https://coreyms.com/').text
with open('corey_blog.html', 'r') as webpage:
    soup = BeautifulSoup(webpage, 'lxml')

"""
Structure de mon objet:
    Pour chaque article: un titre, du contenu, un lien vidéo
"""
titles = []
contents = []
videos = []

for article in soup.find_all('article'):
    titles.append(article.a.text)
    contents.append(article.div.p.text)
    try:
        vid = article.find('iframe', class_='youtube-player')['src']
        vid_id = vid.split('/')[4].split('?')[0]
        link = "www.youtube.com/watch?=" + vid_id
        videos.append(link)
    except Exception:
        link = None
        videos.append(link)
```

```python
# Constructing the articles dictionary
articles_dict = {'title':titles, 'content':contents, 'video':videos}
```

```python
# Constructing the articles dataframe
articles_df = pd.DataFrame.from_dict(articles_dict)
articles_df
```
